## e. Run the query using Hadoop and Spark for 5 times and plot the graph in comparing both methods (running time vs iteration). Results shown below do not contain correct values. It is only to show the format.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Excel file
df = pd.read_excel('/content/QueryExecutuionStatistics.xlsx')\

# Extract data from DataFrame
iterations = df['Iteration'].tolist()
apache_spark = df['Year wise carrier delay time (ms) Apache Spark'].tolist()
map_reduce = df['Year wise carrier delay time (ms) MapReduce'].tolist()

# Width of each bar
bar_width = 0.35

# Index for the x-axis
index = np.arange(len(iterations))

# Plotting the grouped bar chart
plt.bar(index, apache_spark, bar_width, label='Apache Spark', color='#E69BDB')
plt.bar(index + bar_width, map_reduce, bar_width, label='MapReduce', color='#9BC0E6')

# Adding labels and title
plt.xlabel('Iteration')
plt.ylabel('Query execution time (ms)')
plt.title('Year wise carrier delay query execution time vs Iteration')
plt.xticks(index + bar_width / 2, iterations)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
